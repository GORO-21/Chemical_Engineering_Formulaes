```python
import numpy as np
import matplotlib.pyplot as plt

def time_for_conversion(X, k): # To calculate the time it takes to reach the given conversion (min)
    """ X = conversion 
        k = rate constant (min^-1 for first order)
        returns
        t = time required to reach that conversion (min)"""
    return np.log(1 - X) / -k

def concentration_after_conversion (Ca0,X): # Concentration of reactant after conversion
    """ Ca0 = initial concentration (mol/m^3)
        X = Conversion """
    return Ca0*(1-X)

def concentration_vs_time (Ca0,k,t):
    return Ca0 * np.exp((-k)*t)
    
X = 0.9
k = 0.05

Ca0 = 2 # mol/m^3

t = time_for_conversion(X, k)
Ca = concentration_after_conversion (Ca0,X)



print (f"t: {t:.4f} min for {X} conversion")
print (f"Ca: {Ca:.4f} mol/m^3")



```

```python
#Problem data 
Ca0 = 2
k = 0.05
t_max = 100

# generate time plot

t = np.linspace(0,t_max, 100)

Ca = concentration_vs_time (Ca0,k,t)

plt.figure(figsize=(8, 5))
plt.plot(t, Ca, label='Concentration $C_A$ (mol/L)', color='blue')
plt.xlabel('Time (min)')
plt.ylabel('Concentration (mol/L)')
plt.title('Concentration vs. Time in Ideal Batch Reactor')
plt.grid(True)
plt.legend()
plt.show()
```
